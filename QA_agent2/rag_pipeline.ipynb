### Load First Dataset

```python
import json

def load_whats_cooking(path="train.json"):
    with open(path, "r") as f:
        data = json.load(f)

    docs = []
    for r in data:
        cuisine = r["cuisine"]
        ingredients = ", ".join(r["ingredients"])
        doc = f"""
Cuisine: {cuisine}
Ingredients: {ingredients}
"""
        docs.append(doc)

    return docs
```
